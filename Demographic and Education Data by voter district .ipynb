```python
import pandas as pd
```

```python
# import data from https://factfinder.census.gov/
data = pd.read_csv('ACS_17_5YR_DP05.csv')
```

```python
data
```

```python
data.shape
```

```python
# change the names of the columns to the values of the first row
data.columns = data.iloc[0]
```

```python
data.head()
```

```python
#drop the first row
data = data.reindex(data.index.drop(0))
```

```python
data.head()
```

```python
# create a new dataframe that only contains the columns that contain "Margin of Error"
df = data.filter(like='Margin of Error')
```

```python
# get a list of all the column names that contain "Margin of Error"
names = list(df.columns)
```

```python
# check the list
names
```

```python
# drop all the columns that contain "Margin of Error" from the original dataframe
data = data.drop(columns=names)
```

```python
data
# this removed 178 columns from our dataset
```

```python
# drop columns were all the data is NaN
df = data.dropna(how='all', axis='columns')
```
